```python
import glob
import numpy
import matplotlib
import matplotlib.pyplot as plt
from numpy import genfromtxt


#reading files
files = sorted(glob.glob("data*.csv"))

def visualize(file):

    ## reading csv file in array
    my_data = genfromtxt(f'{file}', delimiter=',')

    ## storing all data

    x = []
    mi = []
    ma = []
    av = []

    for c in range(0,40):
        value = my_data[:,c]

        ## checking min,max,avg
        minn = min(value)
        maxx = max(value)
        averagee = int(sum(value)/len(value))

        x.append(c)
        mi.append(minn)
        ma.append(maxx)
        av.append(averagee)

    ## ploting all 3 graphs
    fig, axs = plt.subplots(1,3)
    axs[0].plot(x, av)
    axs[0].set_title(f"{file}")
    axs[0].set_xlabel('Features')
    axs[0].set_ylabel('Average')
    axs[1].plot(x, ma)
    axs[1].set_xlabel('Features')
    axs[1].set_ylabel('Max')
    axs[2].plot(x, mi)
    axs[2].set_xlabel('Features')
    axs[2].set_ylabel('Min')

    plt.show()

def identify_issues(file):

    ## reading csv file in array
    my_data = genfromtxt(f'{file}', delimiter=',')
    
    ## calculating max value
    zero = max(my_data[:,0])
    twenty = max(my_data[:,20])
    
    ## conditions for identify_issues
    if zero == 0 and twenty ==  20:
        print(f"{file}: I doubt the integrity of this dataset")
    else:
        flag = True

        for c in range(0,40):
            if min(my_data[:,c] == 0):
                flag = True
            else:
                flag = False
        
        if flag:
            print(f"{file}: Why is this happening?")
        else:
            print(f"{file}: Dataset looks good")


for file in files[:4]:
    #calling both  functions
    visualize(file)
    identify_issues(file)

    
```
